```python
import os
import struct
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open("eddy1.dat", 'rb')
```

```python
n = int(13E6)

datai = np.zeros((n))
dataq = np.zeros((n))
for i in range(n):
    datai[i] = struct.unpack('f', file.read(4))[0]
    dataq[i] = struct.unpack('f', file.read(4))[0]
```

```python
plt.rcParams["figure.figsize"] = (20,3)
plt.plot(datai)
plt.plot(dataq)
plt.show()
```

```python

sp = 9963145
syms = 380
datai = datai[sp:sp+syms*13]
dataq = dataq[sp:sp+syms*13]

plt.rcParams["figure.figsize"] = (20,3)
plt.plot(datai)
plt.plot(dataq)
plt.show()
```

```python
size = np.size(dataq)
file_q = open ("dataq.txt", "w+")
for i in range(size):
    file_q.write("%f\n" % dataq[i])
file_q.close()
```

```python
file_i = open ("datai.txt", "w+")
for i in range(size):
    file_i.write("%f\n" % datai[i])
file_i.close()
```
